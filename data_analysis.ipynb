```python
import pandas as pd

# Load your CSV
df = pd.read_csv("processed_conversations.csv")

# Preview the data
df.head()
```

```python
import string

# Simple German stopwords list (can be extended)
stopwords = set("""
aber als am an auch auf aus bei bin bis bist da dadurch daher darum das da√ü dass dein deine dem den der des dessen deshalb die dies dieser dieses doch dort du durch ein eine einem einen einer eines er es euch f√ºr hatte haben ich ihm ihn ihnen ihr ihre im in ist ja jede jedem jeden jedoch jene jener jenes jetzt kann kein keine meinem meinen meiner meines mich mir mit muss m√ºssen nach nicht nichts noch nun oder ohne sehr sein seine sich sie sind so solche solcher solches soll sollen sollte ...
""".split())

# Preprocessing function
# convert text to lowercase
# removes punctuation
# remove German stopwords
def preprocess(text):
    text = str(text).lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    words = [word for word in text.split() if word not in stopwords]
    return ' '.join(words)

# Apply it
df['cleaned_message'] = df['user_message'].apply(preprocess)
```

```python
df.head()
```

```python
#converting text to features

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_df=0.95, min_df=2)
# Converts the text data into a document-term matrix (dtm).
# The result, dtm, is a sparse matrix where:
# Rows = documents (each message)
# Columns = words (terms)
# Values = count of how many times each word appears in each document.

dtm = vectorizer.fit_transform(df['cleaned_message'])

```

```python
# Apply topic modelling LDA is a topic modelling algorithm

from sklearn.decomposition import LatentDirichletAllocation

# tells algorithm to find 5 topics in the dtm created above

lda = LatentDirichletAllocation(n_components=5, random_state=42)
lda.fit(dtm)


# Show top words per topic
def display_topics(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"\nIntent {topic_idx+1}:")
        print(", ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]))

display_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
# Visualize topic distribution
import matplotlib.pyplot as plt

# Count dominant topic for each message
topic_distribution = lda.transform(dtm)
dominant_topics = topic_distribution.argmax(axis=1)


df['intent'] = dominant_topics
```

```python
# Assuming your existing value counts
counts = df['intent'].value_counts().sort_index()



# Sort values from highest to lowest
counts = counts.sort_values(ascending=False)

# Plot horizontal bar chart
ax = counts.plot(kind='barh', color='green', figsize=(8, 5))

# Add value labels to each bar
for index, value in enumerate(counts):
    plt.text(value + 0.1, index, str(value), va='center')
    
plt.gca().invert_yaxis()

# Labels and title
plt.title("Distribution of Inferred Keyword Groupings", fontweight='bold')
plt.xlabel("Number of Messages")
plt.ylabel("Keyword groupings")

plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
